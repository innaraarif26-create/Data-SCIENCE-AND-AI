```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
xorData=pd.DataFrame()
xorData ['input 1']=[1,1,0,0]
xorData ['input 2']=[1,0,1,0]
xorData ['output']=[0,1,1,0]

xorData
```

```python
sns.scatterplot(x= xorData['input 1'],y= xorData['input 2'],hue=xorData['output'],s=200)
```

```python
from sklearn.linear_model import Perceptron
clf= Perceptron()
clf.fit(xorData[['input 1','input 2']],xorData['output'])
```

```python
clf.coef_
```

```python
clf.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(xorData.iloc[:,0:2].values,xorData.iloc[:,-1].values,clf=clf,legend=2)
```
