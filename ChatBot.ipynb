```python
%%writefile app.py
import streamlit as st
from transformers import pipeline
import os

# Set Hugging Face token (optional if using public models)
os.environ["HUGGINGFACEHUB_API_TOKEN"] = "your token here"

# Page config
st.set_page_config(page_title="AI Customer Support Chatbot", page_icon="ðŸ¤–", layout="wide")

# Sidebar
with st.sidebar:
    st.image("https://cdn-icons-png.flaticon.com/512/4712/4712027.png", width=120)
    st.header("Company Info")
    st.write("**Services:** AI, Automation, Software Development")
    st.write("**Hours:** Monâ€“Fri, 9 AM â€“ 6 PM")
    st.write("**Email:** support@company.com")

# Title
st.title("ðŸ¤– AI Customer Support Chatbot")
st.markdown("Welcome! I am here to help you with your questions. Ask me anything about our services or AI!")

# Hard-coded system prompt
SYSTEM_PROMPT = """
You are an AI customer support assistant.
Answer politely and professionally.
If you don't know the answer, suggest contacting human support.
"""

# Initialize GPT-2 generator
generator = pipeline("text-generation", model="gpt2")

# Initialize chat state
if "messages" not in st.session_state:
    st.session_state.messages = [
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "assistant", "content": "Hello ðŸ‘‹ How can I help you today?"}
    ]

# User input
user_input = st.chat_input("Type your message here...")

if user_input:
    st.session_state.messages.append({"role": "user", "content": user_input})

    # --- Hybrid response logic ---
    user_lower = user_input.lower()
    if "service" in user_lower:
        reply = "We offer AI, Automation, and Software Development services."
    elif "hours" in user_lower or "time" in user_lower:
        reply = "Our business hours are Monday to Friday, 9 AM to 6 PM."
    elif "contact" in user_lower or "email" in user_lower:
        reply = "You can reach us at support@company.com."
    elif "price" in user_lower or "cost" in user_lower:
        reply = "Pricing depends on the service. Please contact support@company.com for details."
    else:
        # fallback to GPT-2 for casual/open-ended questions
        prompt = SYSTEM_PROMPT + "\nUser: " + user_input + "\nAssistant:"
        hf_response = generator(prompt, max_length=100, do_sample=True, temperature=0.7)[0]['generated_text']
        reply = hf_response.split("Assistant:")[-1].strip()

    st.session_state.messages.append({"role": "assistant", "content": reply})

# Display chat
for msg in st.session_state.messages:
    if msg["role"] != "system":
        with st.chat_message(msg["role"]):
            st.markdown(msg["content"])
```
