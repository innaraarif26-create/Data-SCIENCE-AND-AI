```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
X_train[0]
```

```python
y_train
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[0])
```

as the vaules in the images are between 0 to 255 ,they need to be conveted between 0 and 1 to converge faster(similar range ) so

```python
X_train=X_train/255
X_test= X_test/255

X_train
```

```python
model =Sequential()
```

now the problem is that we have the data in  24 * 24  format,but the architecture we are following must have 784 inputs,so to convert that we will use flatten layer.

```python
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))

model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=10,validation_split=0.2)
```

```python
yprob= model.predict(X_test)
```

```python
ypred= yprob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,ypred)
```
